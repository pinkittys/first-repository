```python
# 라이브러리 임포트
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 데이터 가져오기
cars_df = pd.read_csv('/aiffel/data/cars.csv')
brand_df = pd.read_csv('/aiffel/data/brand.csv')
```

```python
# 데이터 프레임 cars_df 상위 5개 행 확인
cars_df.head()
```

```python
# title 열을 소문자로 변환
cars_df['title'] = cars_df['title'].str.lower()
```

```python
# cars_df의 title 열을 공백을 기준으로 분할하고 앞의 값만 남기기
cars_df['title'] = cars_df['title'].str.split().str[0]
```

```python
# cars_df 타이틀 확인
cars_df['title']
```

```python
# 데이터 프레임 합치기
data_df = pd.merge(cars_df, brand_df, on='title', how='left')
```

```python
# 원-핫 인코딩
data_df = pd.get_dummies(data_df, columns=['Engine', 'Emission Class'])
```

```python
# 라이브러리 표준화 패키지 임포트
from sklearn.preprocessing import StandardScaler
```

```python
# 표준화 객체화
scaler = StandardScaler()
```

```python
# 데이터 프레임 학습 및 변환
scaler.fit_transform(data_df)
```
